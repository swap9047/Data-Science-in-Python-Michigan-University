# Assignment 2 - Network Connectivity

In this assignment you will go through the process of importing and analyzing an internal email communication network between employees of a mid-sized manufacturing company. 
Each node represents an employee and each directed edge between two nodes represents an individual email. The left node represents the sender and the right node represents the recipient.

```python
import networkx as nx

# This line must be commented out when submitting to the autograder
#!head email_network.txt
```

### Question 1

Using networkx, load up the directed multigraph from `email_network.txt`. Make sure the node names are strings.

*This function should return a directed multigraph networkx graph.*

```python
def answer_one():
    
    # Your Code Here
    G=nx.read_edgelist('email_network.txt',data=[('timestamp',int)],create_using=nx.MultiDiGraph())
    return G# Your Answer Here
```

```python
import matplotlib.pyplot as plt
G=answer_one()
pos = nx.random_layout(G)
nx.draw_networkx(G,pos)
```

### Question 2

How many employees and emails are represented in the graph from Question 1?

*This function should return a tuple (#employees, #emails).*

```python
def answer_two():
        
    # Your Code Here
    G=answer_one()
    num_email=len(G.edges())
    num_user=len(G.nodes())
    
    return (num_user,num_email)# Your Answer Here
```

### Question 4

How many nodes are in the largest (in terms of nodes) weakly connected component?

*This function should return an int.*

```python
def answer_four():
        
    # Your Code Here
    wccs=nx.weakly_connected_components(G)
    return len(max(wccs,key=len))# Your Answer Here
```

### Question 6

```python
def answer_six():
        
    # Your Code Here
    G=answer_one()
    sccg=nx.strongly_connected_component_subgraphs(G)
    G_sc=max(sccg,key=len)
    return G_sc # Your Answer Here
```

### Question 11

Which node in G_sc is connected to the most other nodes by a shortest path of length equal to the diameter of G_sc?

How many nodes are connected to this node?


*This function should return a tuple (name of node, number of satisfied connected nodes).*

```python
G_sc=answer_six()
diam=nx.diameter(G_sc)
nodes_dist=[]
dist_dict=nx.all_pairs_shortest_path_length(G_sc)
for k,v in dist_dict.items():
#     print(k,v)
    nodes_dist.append((sum([1 if x==diam else 0 for _,x in v.items()]),k))
# nodes_dist
sorted(nodes_dist,reverse=True)[0]
# [1 if x==diam else 0 for x in v]

```

```python
def answer_eleven():
        
    # Your Code Here
    G_sc=answer_six()
    diam=nx.diameter(G_sc)
    nodes_dist=[]
    dist_dict=nx.all_pairs_shortest_path_length(G_sc)
    for k,v in dist_dict.items():
    #     print(k,v)
        nodes_dist.append((sum([1 if x==diam else 0 for _,x in v.items()]),k))
    # nodes_dist
    val=sorted(nodes_dist,reverse=True)[0]
    return (val[1],val[0])# Your Answer Here
# answer_eleven()
```

### Question 12

Suppose you want to prevent communication from flowing to the node that you found in the previous question from any node in the center of G_sc, what is the smallest number of nodes you would need to remove from the graph (you're not allowed to remove the node from the previous question or the center nodes)? 

*This function should return an integer.*

```python
removal_nodes=[]
G_sc = answer_six()
center = nx.center(G_sc)
node = answer_eleven()[0]
for i in center:
    removal_nodes.extend(list(nx.minimum_node_cut(G_sc, i, node)))
removal_nodes1=[x for x in removal_nodes if x not in center]
len(set(removal_nodes1))
```

```python
def answer_twelve():
        
    # Your Code Here
    removal_nodes=[]
    G_sc = answer_six()
    center = nx.center(G_sc)
    node = answer_eleven()[0]
    for i in center:
        removal_nodes.extend(list(nx.minimum_node_cut(G_sc, i, node)))
    removal_nodes1=[x for x in removal_nodes if x not in center]
    return len(set(removal_nodes1))
# answer_twelve()
```

### Question 13

Construct an undirected graph G_un using G_sc (you can ignore the attributes).

```python
def answer_thirteen():
        
    # Your Code Here
    G_sc=answer_six()
    undir_graph=G_sc.to_undirected()
    G_un = nx.Graph(undir_graph)
    return G_un# Your Answer Here
# answer_thirteen()
```

### Question 14

What is the transitivity and average clustering coefficient of graph G_un?

*This function should return a tuple (transitivity, avg clustering).*

```python
def answer_fourteen():
        
    # Your Code Here
    G_un=answer_thirteen()
    
    return (nx.transitivity(G_un),nx.average_clustering(G_un))# Your Answer Here
# answer_fourteen()
```
